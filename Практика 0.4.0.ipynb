# **Задача 1:**

Интернет-магазин предлагает следующие условия скидок:

*   Для заказов больше 1000 единиц, клиент получает скидку 5%. Если клиент использует промокод SUPERDISCOUNT, он получает скидку 10% (вместо 5%).
*  Для заказов более 5000 единиц, клиент получает скидку 15%, а использование промокода SUPERDISCOUNT увеличивает скидку до 20% (вместо 15%).

Этап 1:
Ввод:
```
Введите стоимость единицы товара: 5
Введите количество товара: 1001
Введите промокод: GiVEMEDISCONT
```

Вывод:

```
Ваша скидка: 5%
Итоговая сумма: 4754.75
```
Этап 2:

Оформите ваш код в виде функции

````python
def get_discount(price: float, count: int, promo_code: str):
    if count > 5000 and promo_code == 'SUPERDISCOUNT':
        return 0.2

    if count > 5000:
        return 0.15

    if count > 1000 and promo_code == 'SUPERDISCOUNT':
        return 0.1

    if count > 1000:
        return 0.05

    return 0

price = float(input('Введите стоимость единицы товара: '))
count = int(input('Введите количество товара: '))
promo_code = input('Введите промокод: ')

discount = get_discount(price, count, promo_code)
total_sum = price * count

print('Ваша скидка:', 100 * discount)
print('Итоговая сумма:', total_sum * (1 - discount))
````

# **Задача 2:**


Этап 1:
Напишите программу способную отфильтровать список и вывести только положительные элементы


Ввод:
```
-1 5 1 2 -3
```

Вывод:

```
5 1 2
```

Этап 2:

Оформите ваш код в виде функции

````python

def custom_filter(iterable: list, condition):
    for i in iterable:
        if condition(i):
            yield i

numbers = [int(i) for i in input().split()]

print(' '.join([str(i) for i in custom_filter(numbers, lambda x: x > 0)]))
````

# **Задача 3:**



Этап 1:
Напишите программу реализующую Алгоритм Евклида


> Алгоритм Евклида – это алгоритм нахождения наибольшего общего делителя (НОД) пары целых чисел.



Ввод:
```
30 18
```

Вывод:

```
6
```

Этап 2:
Оформите ваш код в виде функции


````python
def alg(x1, x2):
    min_x = None
    if x1 < x2:
        min_x = x1
    else:
        min_x = x2
    for i in range(min_x, 0, -1):
        if x1 % i == 0 and x2 % i == 0:
            return i
    return None
x1, x2 = [int(i) for i in input().split()]
print(alg(x1, x2))
````

# **Задача 4:**

Этап 1:
Напишите функцию программу, которая принимает строку и возвращает список слов и количество их упомнинаний в предложении

Этап 2:
Оформите ваш код в виде функции

Ввод:
```
apple banana apple
```

Вывод:

```
apple: 2,
banana: 1
```

````python
def check_words(text: str):
    words = text.split()
    counts = {}
    for i in words:
        if not counts.get(i):
            counts[i] = 1
        else:
            counts[i] += 1
    return counts

text = input()
counts = check_words(text)
for key in counts:
    print(key + ':', counts[key])
````

Детектор анаграмм Напишите программу на Python, которая принимает в качестве входных данных две строки и проверяет, являются ли они анаграммами друг друга

Этап 2:
Оформите ваш код в виде функции

Ввод:
```
listen, silent
```

Вывод:

```
True
```

````python
def check_anogramm(text1, text2):
    if len(text1) != len(text2):
        return False
    for i in text1:
        if i not in text2:
            return False
    for i in text2:
        if i not in text1:
            return False
    return True

text1, text2 = input().split(', ')
print(check_anogramm(text1, text2))
````

# **Задача 6:**


Шифр ​​Цезаря

Напишите программу на Python, которая реализует шифр Цезаря, простой метод шифрования, который заменяет каждую букву буквой на фиксированное количество позиций вниз по алфавиту. Программа должна запрашивать у пользователя сообщение и значение сдвига, а затем шифровать и расшифровывать сообщение.

Этап 1:

Напишите код для реализации данной задачи

Этап 2:

Оформите код в виде нескольких функций:

* Зашифровывает сообщение
* Расшифровывает сообщение

````python
class Cryptographer:
    def __init__(self, offset: int):
        self.offset = offset
        self.letters = list('абвгдеёжзийклмнопрстуфхцчшщъыьэюя')

    def encrypt(self, message: str):
        return ''.join([self.__move_letter(i, self.offset) for i in message])

    def decrypt(self, message: str):
        return ''.join([self.__move_letter(i, -self.offset) for i in message])

    def __move_letter(self, letter: str, offset: int):
        capitalized = False
        if letter.lower() not in self.letters:
            raise Exception(f'Letter {letter} is incorrect')
        if letter not in self.letters:
            capitalized = True
        letter = letter.lower()
        index = self.letters.index(letter)
        new_index = index + offset
        if new_index >= len(self.letters):
            new_index -= len(self.letters)
        if new_index < 0:
            new_index += len(self.letters)
        return self.letters[new_index] if not capitalized else self.letters[new_index].capitalize()

c = Cryptographer(int(input()))

message = input()
print(c.encrypt(message))
print(c.decrypt(c.encrypt(message)))
````
