# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

***Нельзя использовать: len(), min(), max(), average(), sum()***

Дано:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
def custom_max(l: list):
    result = float('-inf')
    for i in l:
        if i > result:
            result = i
    return result

def custom_sum(l: list):
    result = 0
    for i in l:
        result += i
    return result

def custom_len(l: list):
    result = 0
    for _ in l:
        result += 1
    return result

def custom_avg(l: list):
    total_sum = custom_sum(l)
    total_len = custom_len(l)
    return total_sum / total_len

l = [int(i) for i in input("Введите последовательность чисел: ").split()]

print("Сумма чисел:", custom_sum(l))
print("Среднее арифм.:", custom_avg(l))
print("Максимальное значение:", custom_max(l))
````

Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Ввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
import random


commands = ['камень', 'ножницы', 'бумага']
win = {
    'камень': 'бумага',
    'ножницы': 'камень',
    'бумага': 'ножницы'
}

messages = {
    0: 'Вы проиграли',
    1: 'Ничья',
    2: 'Вы выиграли'
}

def get_computer_command():
    return commands[random.randint(0, 2)]

def get_person_result(computer_command, person_command):
    if computer_command == person_command:
        return 1
    if win[computer_command] == person_command:
        return 2
    return 0


computer_command = get_computer_command()
person_command = input("Введите свой ход: ")
result_1 = get_person_result(computer_command, person_command)
print("Компьютер показал " + computer_command + ". " + messages[get_person_result(computer_command, person_command)])

computer_command = get_computer_command()
person_command = input("Введите свой ход: ")
result_2 = get_person_result(computer_command, person_command)
print("Компьютер показал " + computer_command + ". " + messages[get_person_result(computer_command, person_command)])

computer_command = get_computer_command()
person_command = input("Введите свой ход: ")
result_3 = get_person_result(computer_command, person_command)
print("Компьютер показал " + computer_command + ". " + messages[get_person_result(computer_command, person_command)])

def get_total_result(*results):
    if sum(results) / len(results) == 1.0:
        return "Ничья"
    if sum(results) / len(results) > 1.0:
        return "Вы выиграли"
    return "Вы проиграли"

print(get_total_result(result_1, result_2, result_3))
````

# **Задача 5.**

Создайте игру "Палач" или "Виселица", в которой пользователь должен угадать слово, предлагая буквы. Программа должна выбрать случайное слово, а затем разрешить пользователю угадывать буквы. После каждого угадывания программа должна сообщать пользователю, есть ли в слове буква или нет.

**Этап 1**
Создать саму игру

**Этап 2:**
Ввести подсчет количества очков и ограничения на попытки

**Этап 3:** Добавить возможность сразу угадать слово

````python
import random

words = 'создать игру палач виселица которой пользователь должен угадать слово предлагая буквы'.split()
word = words[random.randint(0, len(words) - 1)]

result = False
attempts_count = 0
attempts_border = len(word)
points_count = 0

print(word)

letters = [None for _ in word]

while None in letters and attempts_border > attempts_count:
    letter = input().lower()
    if len(letter) > 1:
        if letter != word:
            print("К сожалению, вы не угадали слово")
        else:
            letters = [i for i in word]
            break
    if letter in letters:
        print('Такая буква уже была введена')
        continue
    if letter in word:
        for index in [idx for idx, val in enumerate(word) if val == letter]:
            letters[index] = letter
            points_count += 1
        print(' '.join(['_' if i is None else f'{i}' for i in letters]))
    else:
        print('Этой буквы нет в слове( Попробуйте еще раз')
    attempts_count += 1

if None in letters:
    print(f"К сожалению, количество попыток кончилось. Количество очков: {points_count}. Слово: " + word)
else:
    print("Вы угадали! Слово: " + word)
````
