# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def find_unique_elements(a, b):
    result = []
    c =  (a + b)
    for i in c:
        if c.count(i) == 1:
            result.append(i)
    return result

print(find_unique_elements([1, 2, 3, 4], [2, 3, 4, 5]))
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
import random


class VirtualPets:
    def __init__(self):
        self.pets = []

    def adoption_pet(self, name):
        self.pets.appnd(VirtualPet(name))

    def show_pet_state(self, name):
        pet = next(i for i in self.pets if i.name == name)
        state = pet.get_state()
        for i in pet.states_params:
            print(i[1] + ':', state[state[i[0]]])


class VirtualPet:
    def __init__(self, name):
        self.name = name
        self.states_params = [('happy', 'Радость'), ('hungred', 'Голод'), ('energy', 'Энергия')]
        food_count = 0

    def feed(self):
        if self.food_count < 10:
            self.food_count += 1
        else:
            print('Переполнение питомца')

    def get_state(self):
        state = {}
        state['happy'] = self.food_count
        state['energy'] = self.food_count
        state['hungred'] = 10 - self.food_count
        return state


def main():
    pets = VirtualPets()

    reqs = {
        'Приручить': lambda name: pets.apode_pet(name),
        'Покормить': lambda name: pets.feed_pet(name),
        'Статус питомца': lambda name: pets.show_pet_state(name),
    }

    while True:
        req = input('Введите действие')
        if req == 'Выйти':
            break
        name = input('Введите имя питомца')
        reqs[req](name)


if __name__ == '__main__':
    main()
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
import random


weapons: {
    'Меч': lambda _: list(range(5, 11)),
    'Лазер': lambda _: list(range(15, 31)),
    'Пистолет': lambda _: list(range(10, 16)),
    'Кинжал':  lambda _: list(range(3, 9))
}


class Warrior:
    def __init__(self, name, weapon):
        self.name = name
        self.weapon = weapon
        self.health = 1000

    def get_damage(self):
        return 1000 - self.health

    def strike(self, d):
        self.health -= d

    def heal(self, h):
        if self.health + h > 1000:
            return
        self.health += h


class Dragon:
    def __init__(self):
        self.weapon = weapons[list(weapons.keys())[random.randint(0, 3)]]
        self.health = 10000

    def get_damage(self):
        return 10000 - self.health

    def strike(self, d):
        self.health -= d


class FightingSystem:
    def __init__(self):
        self.dragon = None
        self.warrior = None

    def create_warrior(self, name, weapon):
        self.warrior = Warrior(name, weapon)

    def create_dragon(self):
        self.warrior = Dragon(name, weapon)

    def heal_warrior(self, health):
        self.warrior.heal(health)

    def strike_dragon(self):
        self.dragon.strike(weapons[self.warrior.weapon]['damage']())

    def strike_warrior(self):
        self.warrior.strike(weapons[self.dragon.weapon]['damage']())


def main():
    system = FightingSystem()

    wname = input('Имя рыцаря: ')
    wweapon = input('Оружие (выберете из ' + ', '.join(list(weapons.keys())) + ')')

    system.create_warrior(wname, wweapon)
    system.create_dragon()

    while True:
        if system.dragon.health == 0:
            print('Warrior won')
            break

        if system.dragon.health == 0:
            print('Dragon won')
            break

        action = input('Уклониться или подлечиться или нанести удар (У/П/НУ)?')

        if action == 'У':
            continue

        if action == 'П':
            warrior_damage = system.heal_warrior(100)
            warrior_damage = system.strike_warrior()
            print('Вам нанемен урон', warrior_damage)

        if action == 'НУ':
            dragon_damage = system.strike_dragon()
            print('Вы нанесли урон', dragon_damage)
            warrior_damage = system.strike_warrior()
            print('Вам нанемен урон', warrior_damage)


if __name__ == '__main__':
    main()
````

# Задание 3



Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests


class Pokemons:
    def __init__(self):
        self.api_pokemons = self.get_pokemons()
        self.team = []

    def get_pokemons(self):
        res = requests.get('https://pokeapi.co/api/v2/pokemon?limit=200')
        if res.status_code != 200:
            return []
        return res.json().get('results')

    def add_pokemon_to_team(self, name):
        pokemon = next(i for i in self.api_pokemons if i.get('name') == name)
        existed_pokemon = self.get_pokemon_by_name(name)
        if existed_pokemon:
            return
        res = requests.get(pokemon.get('url'))
        if res.status_code != 200:
            return
        self.team.append(res.json())

    def remove_from_team(self, name):
        existed_pokemon = self.get_pokemon_by_name(name)
        if not existed_pokemon:
            return
        del self.team[self.team.index(existed_pokemon)]

    def get_pokemon_by_name(self, name):
        print(self.team)
        if len(self.team) == 0:
            return None
        return next(i for i in self.team if i.get('name') == name)

    def show_pokemons_team(self):
        for i in self.team:
            print(i)

    def show_pokemon_info(self, name):
        print(self.get_pokemon_by_name(name))


def main():
    pokemons = Pokemons()
    while True:
        s = input("Введите команду")
        if s == 'exit':
            break
        print(', '.join([i.get('name') for i in pokemons.api_pokemons]))
        pname = input('Введите имя покемона')
        if s == 'Добавить покемона':
            if pname not in [i.get('name') for i in pokemons.api_pokemons]:
                print('Покемон с таким именем не загружен')
                continue
            pokemons.add_pokemon_to_team(pname)
            continue
        if s == 'Удалить покемона':
            if not pokemons.get_pokemon_by_name(pname):
                print('Покемон с таким именем не нфден в команде')
                continue
            pokemons.remove_from_team(pname)
            continue
        if s == 'Показать информацию о покемоне':
            if not pokemons.get_pokemon_by_name(pname):
                print('Покемон с таким именем не нфден в команде')
                continue
            pokemons.show_pokemon_info(pname)
            continue
        if s == 'Тренировочный бой':
            p2name = input()
            if not pokemons.get_pokemon_by_name(p2name):
                print('Покемон с таким именем не нфден в команде')
                continue
            print(pname + ' strike ' + p2name)


if __name__ == '__main__':
    main()
````
