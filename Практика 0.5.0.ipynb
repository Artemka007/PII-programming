# **Задание 2**




Дан список, состоящий из целых чисел, необходимо написать функцию считающую сумму всех положительных четных чисел списка

Ввод:

```
1, 2, 3, 4, 5, 6, 7, 8, 9
```

Вывод:


```
20
```

***Запрещено:***

*   Использование готовых функций для суммирования чисел


````python
def custom_sum(numbers):
    s = {0: 0, 1: 0}
    for i in numbers:
        if i > 0:
            s[i % 2] += i
    return s[0]

print(custom_sum([int(i) for i in input().split(', ')]))
````

# **Задание 5**

Дан список слов, неограниченной длинны, сформируйте словарь, где в качестве ключа будет слово, а в качестве значения количество слов

**Критерии**


*   Словарь необходимо отсортировать по убыванию количества элементов в списке.
*   Подсчет элементов должен быть реализован в отдельной функции
*   Сортировка пары `ключ:значение` должна быть реализована также в виде отдельной функции




Дано:
```
['apple','banana','orange','apple','apple','banana']
```


Вывод:
```
{'apple':3, 'banana': 2, 'orange': 1}
```

***Запрещено:***

*   Использование готовых функций для сортировки
*   Использование готовых функций для подсчета элементов



````python
def partition(data: dict, low, high):
    keys = list(data.keys())
    values = list(data.values())
    pivot = values[(low + high) // 2]
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while values[i] < pivot:
            i += 1

        j -= 1

        while values[j] > pivot:
            j -= 1

        if i >= j:
            break

        values[i], values[j] = values[j], values[i]
        keys[i], keys[j] = keys[j], keys[i]

    for key in keys:
        del data[key]

    for i in range(0, len(keys)):
        data[keys[i]] = values[i]

    return j



def quick_sort(data):
    def _quick_sort_rec(data, low, high):
        if low < high:
            split_index = partition(data, low, high)
            _quick_sort_rec(data, low, split_index)
            _quick_sort_rec(data, split_index + 1, high)

    _quick_sort_rec(data, 0, len(data.values()) - 1)
    return data

def reverse_obj(data):
    keys = list(data)
    for i in range(0, len(data) // 2):
        j = len(data) - 1 - i
        keys[i], keys[j] = keys[j], keys[i]
    return {key: data[key] for key in keys}

def count_objects(data: list):
    counted_data = {}

    for i in data:
        if not counted_data.get(i):
            counted_data[i] = 1
        else:
            counted_data[i] += 1

    return counted_data


l = ['apple','banana','orange','apple','apple','banana', 'abc', 'lob', 'lob', 'abc', 'abc', 'abc']

print(reverse_obj(quick_sort(count_objects(l))))
````

````python
def show_older_than_30(data):
    for i in data:
        if data.get(i).get('age') > 30:
            yield i

def cities(data):
    result = {}
    for i in data:
        city = data.get(i).get('city')
        if not result.get(city):
            result[city] = 1
        else:
            result[city] += 1
    return result

def occupations(data):
    result = {}
    for i in data:
        occupation = data.get(i).get('occupation')
        if not result.get(occupation):
            result[occupation] = [i]
        else:
            result[occupation].append(i)
    return result

print(list(show_older_than_30(people_info)))
print(cities(people_info))
print(occupations(people_info))
````

Задание: Разработка системы отзывов о предметах

Описание: Создать программу на Python для хранения и управления отзывами о предметах учебного курса. Программа должна позволять пользователям добавлять, просматривать и удалять отзывы, а также вычислять средний балл по заданному предмету.

**Функционал:**

*   Добавление отзыва и оценки:
   *   Пользователь может ввести название предмета, оценку (от 1 до 5) и текст отзыва.
   *   Отзывы должны храниться в структуре данных (например, словаре), где ключом будет название предмета, а значением - список отзывов (каждый отзыв может хранить оценку и комментарий).
*   Просмотр отзывов и оценок:
   *   Пользователь может запросить отзывы для указанного предмета.
   *   Если для указанного предмета есть отзывы, программа должна отобразить список всех отзывов и соответствующих оценок.
*   Удаление отзыва:
   *   Пользователь может удалить отзыв по индексу. Необходимо заранее уведомить пользователя о том, какие отзывы доступны для удаления.

*   Вычисление среднего балла по предмету:
   *   Пользователь может ввести название предмета, и программа должна вычислить и вывести средний балл по всем отзывам для этого предмета.
   *   Если отзывов нет, программа должна сообщить об этом.


**Критерии:**

*   Код должен быть оформлен в виде функций
*   Необходимо обрабатывать неправильный ввод пользователя
*   Должны быть комментарии к функциям
*   Присутсвует весь дополнительный функционал



**Опционально:**

Предлагаю вам добавить свои критерии оценки или вопросы, на которые должен ответить студент, чтобы оценить пару




````python
class Comment:
    def __init__(self, name: str, rating: int, message: str):
        if rating < 1 or rating < 5:
            raise ValueError('Оценка должна быть не меньше 1 и не больше 5')
        self.name = name
        self.rating = rating
        self.message = message

    def __str__(self):
        return self.message + ', ' + self.rating

class RatingManager:
    def __init__(self):
        self.comments = []

    def get_comments_by_name(self, name):
        return list(filter(lambda x: x.name == name, self.comments))

    def print_comments_by_name(self, name):
        comments = self.get_comments_by_name(name)
        if len(comments) == 0:
            print('По этому предмету нет отзывов')
            return
        for i in comments:
            print(str(i))

    def create_comment(self, name, rating, message)
        self.comments.append(Comment(name, rating, message))
        print('Отзыв успешно создан')

    def delete_comment(self, index):
        self.comments.pop(index)

    def print_deletable_comments(self):
        print('Вы можете удалить отзывы с идентификаторами:', ', '.join([idx for idx, _ in enumerate(self.comments)]))

    def count_average_rating_by_name(self, name):
        all_ratings = [i.rating for i in self.comments]
        return sum(all_ratings) / len(all_ratings)

    def print_average_rating_by_name(self, name):
        comments = self.get_comments_by_name(name)
        if len(comments) == 0:
            print('По этому предмету нет отзывов')
            return
        print(self.count_average_rating_by_name(name))
````
